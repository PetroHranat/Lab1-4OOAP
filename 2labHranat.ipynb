Варіант No4
Автовиробник пропонує нову модель легкового автомобіля. Дана модель
доступна у 4 стандартних комплектаціях різної вартості. Кожна комплектація
характеризується типом двигуна (бензиновий чи дизельний), об’ємом двигуна,
антиблокувальною гальмівною системою (ABS), електронною системою
стабілізації (ESP), кількістю подушок безпеки, наявністю бортового
комп’ютера, кондиціонером чи клімат-контролем, обшивкою салону. За назвою
комплектації покупець може переглянути її характеристики. Розробити
програму для отримання інформації про конкретну комплектацію автомобіля
та його вартість. Для створення об’єктів скористатись шаблоном Будівельник
(Builder).

```python
class Car:
    def __init__(self):
        self.engine_type = None
        self.engine_capacity = None
        self.abs = None
        self.esp = None
        self.airbags = None
        self.board_computer = None
        self.climate_control = None
        self.interior = None
        self.price = None

    def __str__(self):
        return (f"Двигун: {self.engine_type}, {self.engine_capacity}л\n"
                f"ABS: {'є' if self.abs else 'немає'}\n"
                f"ESP: {'є' if self.esp else 'немає'}\n"
                f"Кількість подушок безпеки: {self.airbags}\n"
                f"Бортовий комп'ютер: {'є' if self.board_computer else 'немає'}\n"
                f"Клімат-контроль: {'є' if self.climate_control else 'немає'}\n"
                f"Обшивка салону: {self.interior}\n"
                f"Ціна: {self.price} USD")


class CarBuilder:
    def __init__(self):
        self.car = Car()

    def set_engine(self, engine_type, engine_capacity):
        self.car.engine_type = engine_type
        self.car.engine_capacity = engine_capacity
        return self

    def set_abs(self, abs_available):
        self.car.abs = abs_available
        return self

    def set_esp(self, esp_available):
        self.car.esp = esp_available
        return self

    def set_airbags(self, airbags):
        self.car.airbags = airbags
        return self

    def set_board_computer(self, board_computer_available):
        self.car.board_computer = board_computer_available
        return self

    def set_climate_control(self, climate_control_available):
        self.car.climate_control = climate_control_available
        return self

    def set_interior(self, interior):
        self.car.interior = interior
        return self

    def set_price(self, price):
        self.car.price = price
        return self

    def build(self):
        return self.car


class CarDirector:
    @staticmethod
    def build_standard():
        return (CarBuilder()
                .set_engine("бензиновий", 1.6)
                .set_abs(True)
                .set_esp(True)
                .set_airbags(2)
                .set_board_computer(False)
                .set_climate_control(False)
                .set_interior("тканина")
                .set_price(20000)
                .build())

    @staticmethod
    def build_premium():
        return (CarBuilder()
                .set_engine("дизельний", 2.0)
                .set_abs(True)
                .set_esp(True)
                .set_airbags(6)
                .set_board_computer(True)
                .set_climate_control(True)
                .set_interior("шкіра")
                .set_price(35000)
                .build())

    @staticmethod
    def build_sport():
        return (CarBuilder()
                .set_engine("бензиновий", 2.5)
                .set_abs(True)
                .set_esp(True)
                .set_airbags(4)
                .set_board_computer(True)
                .set_climate_control(True)
                .set_interior("алькантара")
                .set_price(40000)
                .build())

    @staticmethod
    def build_luxury():
        return (CarBuilder()
                .set_engine("дизельний", 3.0)
                .set_abs(True)
                .set_esp(True)
                .set_airbags(8)
                .set_board_computer(True)
                .set_climate_control(True)
                .set_interior("шкіра преміум")
                .set_price(50000)
                .build())


# Функція для отримання інформації про автомобіль
def get_car_info(option):
    if option == "стандарт":
        return CarDirector.build_standard()
    elif option == "преміум":
        return CarDirector.build_premium()
    elif option == "спорт":
        return CarDirector.build_sport()
    elif option == "люкс":
        return CarDirector.build_luxury()
    else:
        return None


# Основна логіка програми
def main():
    while True:
        print("\nВиберіть комплектацію автомобіля або введіть 'вийти' для завершення:")
        print("Доступні комплектації: стандарт, преміум, спорт, люкс")
        option = input("Введіть комплектацію: ").lower()

        if option == "вийти":
            print("Програма завершена.")
            break

        car = get_car_info(option)
        if car:
            print("\nОсь характеристики вибраної комплектації:")
            print(car)
        else:
            print("Невідома комплектація. Спробуйте ще раз.")

# Запуск програми
if __name__ == "__main__":
    main()
```


Варіант No5
Американська фірма-виробник мобільних телефонів відкрила філіал у
одній з країн сходу. Моделі телефонів, що виготовляються у філіалі ті ж самі,

що і у США, але вартість суттєво відрізняється за рахунок економії на оплаті
праці. Інтернет-магазин, який займається продажем телефонів даної марки, дає
покупцю можливість обрати країну виробника. Після оформлення замовлення
покупець отримує інформацію про термін доставки та вартість телефону. Дані
характеристики залежать від країни виробника. Розробити програму для
покупки телефону через інтернет-магазин. Для створення об’єктів
скористатись шаблоном Абстрактна фабрика(Abstract Factory).

```python
from abc import ABC, abstractmethod

# Абстрактний клас Телефон
class Phone(ABC):
    @abstractmethod
    def get_price(self):
        pass

    @abstractmethod
    def get_delivery_time(self):
        pass

# Реалізація телефонів для різних країн
class USPhone(Phone):
    def get_price(self):
        return 1000  # USD

    def get_delivery_time(self):
        return "7-10 днів"

class EastPhone(Phone):
    def get_price(self):
        return 700  # USD

    def get_delivery_time(self):
        return "14-20 днів"

class EuropeanPhone(Phone):
    def get_price(self):
        return 950  # USD

    def get_delivery_time(self):
        return "5-7 днів"

class IndianPhone(Phone):
    def get_price(self):
        return 600  # USD

    def get_delivery_time(self):
        return "10-15 днів"

class ChinesePhone(Phone):
    def get_price(self):
        return 500  # USD

    def get_delivery_time(self):
        return "7-12 днів"

# Абстрактна фабрика для створення телефонів
class PhoneFactory(ABC):
    @abstractmethod
    def create_phone(self):
        pass

# Фабрики для кожної країни
class USPhoneFactory(PhoneFactory):
    def create_phone(self):
        return USPhone()

class EastPhoneFactory(PhoneFactory):
    def create_phone(self):
        return EastPhone()

class EuropeanPhoneFactory(PhoneFactory):
    def create_phone(self):
        return EuropeanPhone()

class IndianPhoneFactory(PhoneFactory):
    def create_phone(self):
        return IndianPhone()

class ChinesePhoneFactory(PhoneFactory):
    def create_phone(self):
        return ChinesePhone()

# Клас для роботи з магазином
class PhoneStore:
    def __init__(self, factory: PhoneFactory):
        self.phone = factory.create_phone()

    def show_order_details(self):
        print(f"Вартість телефону: {self.phone.get_price()} USD")
        print(f"Термін доставки: {self.phone.get_delivery_time()}")

# Основна функція для симуляції покупки
def order_phone():
    while True:
        print("\nОберіть країну виробника телефону:")
        print("1. США")
        print("2. Східна країна")
        print("3. Європа")
        print("4. Індія")
        print("5. Китай")
        print("6. Вихід")
        choice = input("Ваш вибір (1-6): ")

        if choice == "1":
            factory = USPhoneFactory()
        elif choice == "2":
            factory = EastPhoneFactory()
        elif choice == "3":
            factory = EuropeanPhoneFactory()
        elif choice == "4":
            factory = IndianPhoneFactory()
        elif choice == "5":
            factory = ChinesePhoneFactory()
        elif choice == "6":
            print("Вихід з програми. Дякуємо за використання!")
            break
        else:
            print("Невірний вибір! Спробуйте ще раз.")
            continue

        store = PhoneStore(factory)
        store.show_order_details()

        # Питання про повторний вибір
        change_choice = input("\nБажаєте поміняти вибір? (так/ні): ").strip().lower()
        if change_choice != "так":
            print("Дякуємо за покупку!")
            break

# Виклик програми для замовлення телефону
if __name__ == "__main__":
    order_phone()
```
